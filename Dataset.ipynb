```python
dataset = pd.concat([toxic_dataset_1, toxic_dataset_2, toxic_dataset_3, toxic_dataset_4, toxic_dataset_5, toxic_dataset_6, toxic_dataset_7, toxic_dataset_8])
```

```python
cols_target = ["toxic", "severe toxic", "obscene", "threat", "insult", "identity hate"]
```

```python
target_data = dataset[cols_target]

dataset['score'] =dataset[cols_target].sum(axis=1)
print(dataset['score'].value_counts())
dataset['score'] = dataset['score'] > 0 
dataset['score'] = dataset['score'].astype(int)
print(dataset['score'].value_counts())
```

```python
en_ms = pd.read_csv("/content/gdrive/My Drive/Datasets/Code-switching Cyberbullying Detection System/en-ms.csv")
en_ms.head()
```

```python
english = False
malay = False
code_switching_dataset = []
for index, row in dataset.iterrows():
  for en in en_ms["en"]:
    if en in str(row["message"]).lower().split():
      english = True
      break

  for ms in en_ms["ms"]:
    if ms in str(row["message"]).lower().split():
      malay = True
      break
    
  if english and malay:
    print(row)
    code_switching_dataset.append({'message': row["message"], 'score': row["score"]})

  english = False
  malay = False
```

```python
len(code_switching_dataset)
```

```python
with open('dataset.csv', 'w') as filehandler:
    write = csv.writer(filehandler)
    for listitem in code_switching_dataset:
        # filehandler.write(listitem)
        write.writerow([listitem["message"], listitem["score"]])
  
```
